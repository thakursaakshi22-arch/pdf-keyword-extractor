```python
import PyPDF2
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from wordcloud import WordCloud
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix
)
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('vader_lexicon')
```

```python
# (a) & (b) READ PDF & EXTRACT TEXT
pages_text = []
```

```python
with open('/content/2024-ar-dexon-en.pdf', 'rb') as f:
    reader = PyPDF2.PdfReader(f)
    for i, page in enumerate(reader.pages):
        try:
            text = page.extract_text()
        except:
            text = ""
        pages_text.append(text if text else "")

len(pages_text), pages_text[0][:500]   # check number of pages & small sample
```

```python
# PAGE-LEVEL DATAFRAME (useful reference/backup)
df_pages = pd.DataFrame({
    "page_num": np.arange(1, len(pages_text)+1),
    "page_text": pages_text
})

df_pages.head()
```

```python
# (c) BASIC PREPROCESSING

def basic_preprocess(text):
    # remove excessive newlines & spaces
    text = text.replace("\n", " ")
    text = re.sub(r'\s+', ' ', text)
    text = text.strip()
    return text

raw_text = " ".join(pages_text)
preprocessed_text = basic_preprocess(raw_text)

len(raw_text), len(preprocessed_text)
```

```python
# ==========================
# (d) SENTENCE TOKENIZATION
# ==========================
nltk.download('punkt_tab')
sentences = sent_tokenize(preprocessed_text)

# Remove very short or obviously non-informative sentences
sentences = [s for s in sentences if len(s.split()) > 4]

len(sentences)
```

```python
# ==========================
# (e) SENTENCE-LEVEL DATAFRAME
# ==========================
df_sent = pd.DataFrame({
    "sentence": sentences
})

df_sent.head(10)# (e) SENTENCE-LEVEL DATAFRAME
df_sent = pd.DataFrame({
    "sentence": sentences
})

df_sent.head(10)
```

```python
# ==========================
# CLEANING PER SENTENCE (for tokens & models)
# ==========================
stop_words = set(stopwords.words('english'))
```

```python
def clean_sentence_for_model(text):
    text = text.lower()
    # keep only letters and spaces
    text = re.sub(r'[^a-z\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    # remove stopwords
    tokens = [w for w in text.split() if w not in stop_words and len(w) > 2]
    return " ".join(tokens)
```

```python
df_sent["clean_text"] = df_sent["sentence"].apply(clean_sentence_for_model)
df_sent.head()
```

```python
# ==========================
# (f) SENTIMENT ANALYSIS USING VADER
# ==========================
sia = SentimentIntensityAnalyzer()
```

```python
df_sent["sentiment_score"] = df_sent["sentence"].apply(
    lambda x: sia.polarity_scores(x)["compound"]
)
```

```python
def label_sentiment(score):
    if score >= 0.05:
        return "positive"
    elif score <= -0.05:
        return "negative"
    else:
        return "neutral"

df_sent["sentiment_label"] = df_sent["sentiment_score"].apply(label_sentiment)

df_sent[["sentence", "sentiment_score", "sentiment_label"]].head(10)
```

```python
# Sentiment distribution (numbers & percentage)
sent_counts = df_sent["sentiment_label"].value_counts()
sent_pct = df_sent["sentiment_label"].value_counts(normalize=True) * 100
print("Sentiment counts:\n", sent_counts)
print("\nSentiment percentage (%):\n", sent_pct.round(2))
```

```python
# Bar plot
plt.figure(figsize=(6,4))
sns.barplot(x=sent_counts.index, y=sent_counts.values)
plt.title("Sentence-level Sentiment Distribution")
plt.ylabel("Number of sentences")
plt.xlabel("Sentiment")
plt.tight_layout()
plt.show()
```

```python
# ==========================
# (g) WORD TOKENISATION
# ==========================
all_clean_text = " ".join(df_sent["clean_text"])
tokens = word_tokenize(all_clean_text)
```

```python
# Keep purely alphabetic tokens, remove stopwords again (safety)
tokens = [t for t in tokens if t.isalpha() and t not in stop_words]

len(tokens), tokens[:30]
```

```python
# ==========================
# (h) WORD FREQUENCY
# ==========================
from nltk.probability import FreqDist
```

```python
freq_dist = FreqDist(tokens)
top_30 = freq_dist.most_common(30)
```

```python
freq_df = pd.DataFrame(top_30, columns=["word", "count"])
freq_df.head(10)
```

```python
import matplotlib.pyplot as plt
# Bar plot of top 20 frequent words
plt.figure(figsize=(10,5))
sns.barplot(data=freq_df.head(20), x="word", y="count")
plt.xticks(rotation=60)
plt.title("Top 20 Most Frequent Words in DEXON TECHNOLOGY AR 2024-25")
plt.xlabel("Word")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
# ==========================
# (j) DOCUMENT TERM MATRIX (COUNT VECTORIZER)
# ==========================
count_vect = CountVectorizer(max_features=500)  # can tweak if needed
X_count = count_vect.fit_transform(df_sent["clean_text"])
```

```python
dtm = pd.DataFrame(
    X_count.toarray(),
    columns=count_vect.get_feature_names_out()
)
```

```python
# (k) TF-IDF MATRIX
tfidf_vect = TfidfVectorizer(max_features=300)
X_tfidf = tfidf_vect.fit_transform(df_sent["clean_text"])

tfidf_df = pd.DataFrame(
    X_tfidf.toarray(),
    columns=tfidf_vect.get_feature_names_out()
)
```

```python
tfidf_df.shape, tfidf_df.head()
```

```python
# ==========================
# (l) CLASSIFICATION MODELS ON SENTIMENT
# ==========================

# Target variable
y = df_sent["sentiment_label"]

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_tfidf, y,
    test_size=0.2,
    random_state=42,
    stratify=y
)

X_train.shape, X_test.shape
```

```python
def evaluate_model(model, model_name):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    print(f"\n===== {model_name} =====")
    print("Accuracy:", round(accuracy_score(y_test, y_pred), 4))
    print("\nClassification Report:\n", classification_report(y_test, y_pred))

    cm = confusion_matrix(y_test, y_pred, labels=["negative", "neutral", "positive"])
    plt.figure(figsize=(5,4))
    sns.heatmap(cm, annot=True, fmt="d",
                xticklabels=["negative", "neutral", "positive"],
                yticklabels=["negative", "neutral", "positive"])
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.title(f"Confusion Matrix - {model_name}")
    plt.tight_layout()
    plt.show()
```

```python
# 1) Logistic Regression
log_reg = LogisticRegression(max_iter=1000)
evaluate_model(log_reg, "Logistic Regression")
```

```python
# 2) Decision Tree
dt_clf = DecisionTreeClassifier(random_state=42, max_depth=10)
evaluate_model(dt_clf, "Decision Tree")
```

```python
# 3) Naïve Bayes (Multinomial)
nb_clf = MultinomialNB()
evaluate_model(nb_clf, "Multinomial Naive Bayes")
```
